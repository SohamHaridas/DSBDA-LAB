```python
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
df.columns = columns
```

```python
summary_stats = df.groupby("species").agg(['mean', 'median', 'min', 'max', 'std'])
print("Summary Statistics Grouped by Species:\n", summary_stats)
```

```python
for species in df["species"].unique():
    species_data = df[df["species"] == species].describe()
    print(f"\nStatistics for {species}:\n", species_data)
```

```python
numerical_df = df.select_dtypes(include=[np.number])  # Select only numeric columns

mean_values = numerical_df.mean()
median_values = numerical_df.median()
mode_values = numerical_df.mode().iloc[0]

print("\nMean Values:\n", mean_values)
print("\nMedian Values:\n", median_values)
print("\nMode Values:\n", mode_values)
```

```python
std_values = numerical_df.std()
print("\nStandard Deviation Values:\n", std_values)
```

```python
percentiles = numerical_df.quantile([0.25, 0.5, 0.75])
print("\nPercentile Values:\n", percentiles)
```
